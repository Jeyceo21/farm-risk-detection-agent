```python
import os
import shutil

# Move inner folders up to flatten the dataset structure
base_path = "dataset"

# Identify the immediate subdirectories under base_path (e.g., 'PlantVillage', 'plantvillage')
# These are the folders we want to flatten by moving their contents up.
inner_folders_to_process = [d for d in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, d))]

for inner_folder_name in inner_folders_to_process:
    inner_path = os.path.join(base_path, inner_folder_name)

    # Move all items (sub-subfolders, which are the actual class folders) from inner_path to base_path
    for item_name in os.listdir(inner_path):
        src_item_path = os.path.join(inner_path, item_name)
        dst_item_path = os.path.join(base_path, item_name)

        if os.path.isdir(src_item_path): # Ensure we are only dealing with directories (class folders or containers of class folders)
            if os.path.exists(dst_item_path):
                if os.path.isdir(dst_item_path):
                    # Both source and destination are directories, so merge contents (class folders)
                    print(f"Merging contents from directory '{src_item_path}' into existing directory '{dst_item_path}'")
                    for class_folder_name in os.listdir(src_item_path):
                        src_class_folder_path = os.path.join(src_item_path, class_folder_name)
                        dst_class_folder_path = os.path.join(dst_item_path, class_folder_name)

                        if os.path.isdir(src_class_folder_path):
                            if os.path.exists(dst_class_folder_path):
                                if os.path.isdir(dst_class_folder_path):
                                    # Both are class folders, move files from src_class_folder_path to dst_class_folder_path
                                    print(f"  Merging files from class folder '{src_class_folder_path}' into existing '{dst_class_folder_path}'")
                                    for img_file_name in os.listdir(src_class_folder_path):
                                        src_img_path = os.path.join(src_class_folder_path, img_file_name)
                                        dst_img_path = os.path.join(dst_class_folder_path, img_file_name)
                                        if os.path.isfile(src_img_path):
                                            shutil.move(src_img_path, dst_img_path)
                                    os.rmdir(src_class_folder_path) # Remove the now empty source class folder
                                else:
                                    print(f"Warning: Cannot merge directory '{src_class_folder_path}' with existing file '{dst_class_folder_path}'. Skipping.")
                            else:
                                # Destination class folder does not exist, move the entire class folder
                                print(f"  Moving class folder '{src_class_folder_path}' to '{dst_item_path}'")
                                shutil.move(src_class_folder_path, dst_item_path)
                        elif os.path.isfile(src_class_folder_path):
                            # Move files directly under the 'class_folder_container' if any (unlikely for this dataset structure)
                            print(f"  Moving file '{src_class_folder_path}' to '{dst_item_path}'")
                            shutil.move(src_class_folder_path, dst_item_path)
                    os.rmdir(src_item_path) # Remove the now empty source container folder (e.g., 'dataset/plantvillage/PlantVillage')
                else:
                    print(f"Warning: Cannot merge directory '{src_item_path}' with existing file '{dst_item_path}'. Skipping.")
            else:
                # Destination does not exist, simply move the directory (the container of class folders)
                print(f"Moving folder '{src_item_path}' to '{base_path}'")
                shutil.move(src_item_path, base_path)
        elif os.path.isfile(src_item_path): # If src_item_path is a file directly under inner_path
            print(f"Moving file '{src_item_path}' to '{base_path}'")
            shutil.move(src_item_path, base_path)

    # After moving all contents from this inner_path, remove the now empty inner_path folder
    print(f"Removing empty inner folder {inner_path}")
    shutil.rmtree(inner_path)

print("✅ Dataset structure fixed!")
```

```python
dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "dataset",
    image_size=(128,128),
    batch_size=32
)

class_names = dataset.class_names
print("Classes:", class_names)
```

```python
import tensorflow as tf

image_model = tf.keras.Sequential([
    tf.keras.layers.Rescaling(1./255, input_shape=(128,128,3)),
    tf.keras.layers.Conv2D(32, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(64, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(128, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(len(class_names), activation='softmax')
])

image_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

image_model.summary()
```

```python
image_model.fit(dataset, epochs=2)
```

```python
CROP_RULES = {
      "rice": {"temp": (20, 35), "moisture": 60, "humidity": 70},
          "wheat": {"temp": (10, 25), "moisture": 40, "humidity": 50},
              "cotton": {"temp": (21, 30), "moisture": 35, "humidity": 60},
                  "sugarcane": {"temp": (20, 38), "moisture": 65, "humidity": 65}
                  }
```

```python
import cv2
import json
from google.colab import files

def detect_risk(data, image_path):
    crop = data["crop"]
    rules = CROP_RULES[crop]

    risks = []

    # RULE BASED
    if not (rules["temp"][0] <= data["temperature"] <= rules["temp"][1]):
        risks.append(("Temperature Stress", "HIGH"))

    if data["soil_moisture"] < rules["moisture"]:
        risks.append(("Water Stress", "HIGH"))

    if data["humidity"] < rules["humidity"]:
        risks.append(("Pest Probability", "MEDIUM"))

    if data["ndvi"] < 0.4:
        risks.append(("Nutrient Deficiency", "HIGH"))

    # ML anomaly
    X = np.array([[data["temperature"],
                     data["humidity"],
                     data["soil_moisture"],
                     data["ndvi"]]])

    if model.predict(X)[0] == -1:
        risks.append(("Environmental Anomaly", "HIGH"))

    # IMAGE PREDICTION
    img = cv2.imread(image_path)
    img = cv2.resize(img, (128,128))
    img = np.expand_dims(img, axis=0)

    pred = image_model.predict(img)
    label = class_names[np.argmax(pred)]

    print("Detected Disease:", label)

    if "healthy" not in label.lower():
        risks.append(("Disease Detected", "HIGH"))

    # SCORE
    score_map = {"LOW":1, "MEDIUM":2, "HIGH":3}
    score = sum(score_map[r[1]] for r in risks)

    return {
        "agent": "RiskDetectionAgent",
        "crop": crop,
        "risks": risks,
        "risk_score": score,
        "status": "SAFE" if score == 0 else "ALERT"
    }
```

```python
import os
import random

def get_random_image(dataset_path="dataset"):
    all_images = []
    for root, dirs, files in os.walk(dataset_path):
        for file in files:
            if file.endswith((".jpg", ".png", ".jpeg")):
                all_images.append(os.path.join(root, file))
    if not all_images:
        raise FileNotFoundError(f"No images found in the dataset path: {dataset_path}")
    return random.choice(all_images)

# Get random image
img_path = get_random_image()

print("📸 Using image:", img_path)
```

```python
import random

def generate_random_input():
    return {
        "crop": random.choice(["rice", "wheat", "cotton", "sugarcane"]),
        "temperature": random.uniform(5, 45),
        "humidity": random.uniform(30, 90),
        "soil_moisture": random.uniform(20, 80),
        "ndvi": random.uniform(0.2, 0.9)
    }

test_input = generate_random_input()

print("📊 Random Input:", test_input)
```

```python
test_input = generate_random_input()

img_path = get_random_image()

print("📊 Input:", test_input)
print("📸 Image:", img_path)

result = detect_risk(test_input, img_path)

print("\n🚀 FINAL OUTPUT:")
print(json.dumps(result, indent=4))
```
